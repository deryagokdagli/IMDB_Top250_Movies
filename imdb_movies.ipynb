```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_movies = pd.read_csv("IMDBTop250Movies.csv", sep=",", encoding="latin1")
```

```python
df_movies
```

```python
df_movies.drop(["tagline"],axis=1,inplace=True)
```

```python
df_movies.head()
```

```python
df_movies.tail()
```

```python
df_movies["budget"] = df_movies["budget"].replace("Not Available", np.nan)
df_movies["box_office"] = df_movies["box_office"].replace("Not Available", np.nan)
```

```python
df_movies
```

```python
df_movies.isnull().sum()
```

```python
type("box_office")
```

```python
df_movies["budget"] = pd.to_numeric(df_movies["budget"], errors='coerce')
df_movies["box_office"] = pd.to_numeric(df_movies["box_office"], errors='coerce')
```

```python
df_movies["budget"] = df_movies["budget"].fillna(df_movies["budget"].median())
df_movies["box_office"] = df_movies["box_office"].fillna(df_movies["box_office"].median())
```

```python
df_movies.describe()
```

```python
df_movies["genre"]
```

```python
genre_counts = df_movies["genre"].value_counts() 
```

```python
genre_counts[genre_counts > 4]
```

```python
threshold = 4
filtered = genre_counts[genre_counts > threshold]
plt.figure(figsize=(10,6), dpi=200)
filtered.plot(kind="bar", color= "#4A90E2", label="Film sayısı")
plt.title("Top 250 Filmler: Film Türlerine Göre Dağılım (N > 4)",fontsize=14)
plt.xlabel("Türler", fontsize=12)
plt.ylabel("Toplam Film Sayısı", fontsize=12, labelpad=10)
plt.xticks(rotation=20,ha="right")
plt.grid(alpha=0.2)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
rating_counts = df_movies["rating"].value_counts()
```

```python
rating_counts
```

```python
filtered_rating_counts = rating_counts[rating_counts.index >= 8.0].sort_index()
```

```python
filtered_rating_counts
```

```python
plt.figure(figsize=(10,6),dpi=200)
filtered_rating_counts.plot(kind="bar", color="#2E8B57", label= "Film sayısı")
plt.title("Top 250 Filmler: IMDb 8.0 ve Üzeri", fontsize=14)
plt.xlabel("IMDb Puanı", fontsize=12,labelpad=10)
plt.ylabel("Toplam Film Sayısı", fontsize=12, labelpad=10)
plt.xticks(rotation=0)
plt.grid(alpha=0.2)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
bins = range(1920, 2030, 5)  # 1920–2025 arası, 5’er yıllık bakıcaz
df_movies['year_group'] = pd.cut(df_movies['year'], bins)
grouped = df_movies.groupby("year_group", observed=True).size()
```

```python
plt.figure(figsize=(10,6),dpi=200)
grouped.plot(kind="bar", color="#FF4500", label="Film sayısı")
plt.title("Top 250 Filmler: 5 Yıllık Dönemlere Göre Film Sayısı", fontsize=14)
plt.xlabel("Yıllar", fontsize=12,labelpad=10)
plt.ylabel("Toplam Film Sayısı", fontsize=12, labelpad=10)
plt.grid(alpha=0.2)
plt.xticks(rotation=30, ha="right")
plt.tight_layout()
plt.legend()
plt.show()
```

```python
sorted_movies = df_movies.sort_values(by="box_office", ascending=False)
top10 = sorted_movies.iloc[:10]
```

```python
plt.figure(figsize=(10,6),dpi=200)
plt.bar(top10["name"],top10["box_office"],color="#ff69b4", label="Gişe geliri")
plt.yscale('log')
plt.xticks(rotation=90)
plt.ylabel("Toplam Gişe Geliri (10^9 milyar dolar)", fontsize=12, labelpad=10)
plt.xlabel("Film Adı", fontsize=12, labelpad=10)
plt.title("IMDb Top 250: En Çok Kazanan 10 Film", fontsize=14)
plt.xticks(rotation=30, ha="right")
plt.grid(alpha=0.2)
plt.legend()
plt.tight_layout()
plt.show() 
```

```python
df_movies['profit'] = df_movies['box_office'] - df_movies['budget']
```

```python
# En karlı 10 film
top_profit = df_movies.nlargest(10, 'profit')[['name', 'budget', 'box_office', 'profit']]

# En zarar eden 10 film
top_loss = df_movies.nsmallest(10, 'profit')[['name', 'budget', 'box_office', 'profit']]
```
